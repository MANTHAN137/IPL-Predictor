```python
TotalScore=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
TotalScore=TotalScore[TotalScore['inning']==1]
print(TotalScore)
```

```python
 Match_Data=match.merge(TotalScore[['match_id','total_runs']],left_on='id',right_on='match_id')
 Match_Data
```

```python
Match_Data['team1'].unique()
```

```python
Match_Data['team1']=Match_Data['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
Match_Data['team2']=Match_Data['team2'].str.replace('Delhi Daredevils','Delhi Capitals')

Match_Data['team1']=Match_Data['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
Match_Data['team2']=Match_Data['team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

Match_Data['team1']=Match_Data['team1'].str.replace('Gujarat Lions','Gujarat Titans')
Match_Data['team2']=Match_Data['team1'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
WorkingTeams=['Sunrisers Hyderabad', 'Mumbai Indians', 'Gujarat Lions', 'Royal Challengers Bangalore','Kolkata Knight Riders', 'Kings XI Punjab','Chennai Super Kings', 'Rajasthan Royals','Delhi Capitals']
```

```python
Match_Data=Match_Data[Match_Data['team1'].isin(WorkingTeams)]
Match_Data['team1']=Match_Data['team2'].str.replace('Gujarat Lions','Gujarat Titans')
```

```python
Match_Data.info()
```

```python
Match_Data['dl_applied'].value_counts()
```

```python
Match_Data=Match_Data[Match_Data['dl_applied']==0]
```

```python
Match_Data=Match_Data[['match_id','city','total_runs','winner']]
Match_Data
```

```python
delivery.head()
```

```python
delivery=delivery[delivery['inning']==2]
```

```python
Delivery_df=Match_Data.merge(delivery,on='match_id')
```

```python
Delivery_df.shape
```

```python
Delivery_df.head()
```

```python
Delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
Delivery_df['current_score']=Delivery_df.groupby('match_id').cumsum()['total_runs_y']
```

```python
Delivery_df['runs_left']=Delivery_df['total_runs_x']-Delivery_df['current_score']+1
```

```python
Delivery_df.head()
```

```python
Delivery_df['balls_left']=126-(Delivery_df['over']*6+Delivery_df['ball'])
```

```python
Delivery_df['balls_left']
```

```python
Delivery_df['player_dismissed']=Delivery_df['player_dismissed'].fillna("0")
Delivery_df['player_dismissed']=Delivery_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
Delivery_df['player_dismissed']=Delivery_df['player_dismissed'].astype('int')
wickets=Delivery_df.groupby('match_id').cumsum()['player_dismissed'].values
Delivery_df['wickets']=10-wickets
Delivery_df.head()
```

```python
#Current Run rate
Delivery_df['crr']=(Delivery_df['current_score']*6)/(120-Delivery_df['balls_left'])
```

```python
Delivery_df['rrr']=(Delivery_df['runs_left']*6)/(Delivery_df['balls_left'])
```

```python
Delivery_df.info()
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
Delivery_df['result']=Delivery_df.apply(result,axis=1)
```

```python
Final_Dataframe=Delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
Final_Dataframe.head()
```

```python
Final_Dataframe=Final_Dataframe.sample(Final_Dataframe.shape[0]) ##Shuffling to avoid bais outcomes
```
